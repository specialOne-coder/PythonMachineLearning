# 1 - Interpolate


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,10)
y = x**2

plt.scatter(x,y)
plt.show()
```

```python
from scipy.interpolate import interp1d

f = interp1d(x,y,kind='linear')
new_x = np.linspace(1,10,30)
result = f(new_x)

plt.scatter(x,y)
plt.scatter(new_x,result,c='r')
```

# 2 - Optimize

```python
def f(x):
    return x**2 + 15*np.sin(x)
```

```python
x = np.linspace(-10,10,100)
plt.plot(x,f(x))
plt.show()
```

```python
from scipy import optimize

x0=-5
resultat = optimize.minimize(f,x0=x0).x
```

```python
plt.plot(x,f(x),lw=3,zorder=-1)
plt.scatter(resultat,f(resultat),s=100,c='r')
plt.scatter(x0,f(x0),marker='+',s=200,c='g')
```

#  3 - Traitement de signal

detrend permet d'eliminer toute tendance linaire dans un signal

```python
from scipy import signal

x = np.linspace(0,20,100)
y = x + 4*np.sin(x) + np.random.randn(x.shape[0])
plt.plot(x,y)
```

```python
new_y = signal.detrend(y)
plt.plot(x,y)
plt.plot(x,new_y)
```

Transformation de fourier : Technique permettant l'extraction et l'analyse des frquences dans un signal

```python
x = np.linspace(0,30,1000)
y = 3*np.sin(x) + 2*np.sin(5*x) + np.sin(10*x)

plt.plot(x,y)
```

```python
from scipy import fftpack

fourier = np.abs(fftpack.fft(y))
frequence = np.abs(fftpack.fftfreq(y.size))

plt.plot(frequence,fourier)
```
