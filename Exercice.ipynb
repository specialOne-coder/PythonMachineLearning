```python
def fibo(n):
    a = 0
    b = 1
    liste = [a]
    while a < n:
        a,b = b,a+b
        liste.append(a)
        liste.sort(reverse=True)
    return liste
```

```python
print(fibo(100))
```

Generer plusieurs graphiques sur une seule figure:subplot(lignes,col,position)




```python
import numpy as np
dataset = {f"experience{i}":np.random.randn(100) for i in range(4)}
```

```python
import matplotlib.pyplot as plt

def graphique(data):
    n = len(data)
    plt.figure(figsize=(12,8))
    for i,j in zip(data.keys(),range(1, n+1)):
        plt.subplot(n,1,j)
        plt.plot(data[i])
        plt.title(i)
        
    plt.show()
   
    
    
    
```

```python
graphique(dataset)
```

```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x[0]) + np.cos(x[0]+x[1])*np.cos(x[0])
```

```python
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
x,y = np.meshgrid(x,y)
plt.contour(x,y,f(np.array([x,y])),20)

x0 = np.zeros((2,1))
plt.scatter(x0[0], x0[1],marker='+',s=200,c='r')
r = optimize.minimize(f,x0=x0).x
plt.scatter(r[0],r[1],c='g',s=100)
```
