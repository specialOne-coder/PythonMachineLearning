```python
import numpy as np
dataset = {f"experience{i}":np.random.randn(100) for i in range(4)}
```

```python
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12,8))
    for k,i in zip(data.keys(), range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```

```python
graphique(dataset)
```

```python
def graphiquei(ir):
    xi = ir.data
    yi = ir.target
    n = np.unique(yi).size
    plt.figure(figsize=(12,8))
    for ix,iy in zip(xi, range(1,n+1)):
        plt.subplot(n,1,iy)
        plt.scatter(xi[:,iy],xi[:,iy],c=yi,alpha=0.5)
        plt.title(f'Variable{ix}')
    plt.show()
```

```python
graphiquei(iris)
```
