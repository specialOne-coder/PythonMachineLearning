# 1 - Interpolate


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,10)
y = x**2

plt.scatter(x,y)
plt.show()
```

```python
from scipy.interpolate import interp1d

f = interp1d(x,y,kind='linear')
new_x = np.linspace(1,10,30)
result = f(new_x)

plt.scatter(x,y)
plt.scatter(new_x,result,c='r')
```
