# 5 - Scatter

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
xir = iris.data
yir = iris.target
names = list(iris.target_names)
print(f'x contient {xir.shape[0]} exemples et {xir.shape[1]} variables')
print(f'et il y a {np.unique(yir).size} classes')
```

```python
plt.scatter(xir[:,0],xir[:,2],c=yir,alpha=0.5, s=200)
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')
plt.show()
```

```python
import numpy as np
dataset = {f"experience{i}":np.random.randn(100) for i in range(4)}
```

```python
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12,8))
    for k,i in zip(data.keys(), range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```

```python
graphique(dataset)
```

```python
def graphiquei(ir):
    xi = ir.data
    yi = ir.target
    n = np.unique(yi).size
    plt.figure(figsize=(12,8))
    for ix,iy in zip(xi, range(1,n+1)):
        plt.subplot(n,1,iy)
        plt.scatter(xi[:,0],xi[:,iy],c=yi,alpha=0.5,s=100)
        plt.title(iy)
    plt.show()
```

```python
graphiquei(iris)
```
