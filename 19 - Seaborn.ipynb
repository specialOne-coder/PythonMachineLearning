```python
import seaborn as sns
import pandas as pd
import matplotlib as plt
```

```python
iris = pd.read_csv('seaborn-data/iris.csv')
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

```python
titanic = pd.read_csv('seaborn-data/titanic.csv')
titanic = titanic.drop(['sibsp','parch','embarked','class','who','adult_male','deck','embark_town','alive','alone'],axis=1)
```

```python
sns.catplot(x='sex',y='age',data=titanic)
```

```python
sns.catplot(x='pclass',y='age',data=titanic,hue='sex')
```

```python
sns.boxplot(x='pclass',y='age',data=titanic,hue='sex')
```

```python
sns.distplot(titanic['pclass'])
```

```python
sns.jointplot('age','fare',data=titanic,kind='hex')
```

```python
sns.catplot(x='age',y='fare',data=titanic,hue='sex')
```

```python
sns.heatmap(titanic.corr())
```
